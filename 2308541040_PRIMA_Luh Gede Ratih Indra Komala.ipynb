```python
# Jika diketahui bahwa n adalah dua digit terakhir dari NIM kalian, maka tentukan biangan prima antara 1000 * n & 1000 * (n+1)
# NIM = 2308541040, maka 2 digit terakhir dari NIM adalah 40
# Jika diketahui bahwa n adalah 40, maka tentukan biangan prima antara 1000 * n & 1000 * (n+1)

# import math untuk menggunakan sqrt
import math

# Mengecek apakah suatu bilangan prima
def is_prime(bilangan_40):
    # Langkah 1. Pengecekan terhadap kasus sederhana
    if bilangan_40 <= 1:
        return False # bilangan yang kurang dari atau sama dengan 1 bukan bilangan prima
    if bilangan_40 == 2:
        return True  # bilangan 2 adalah bilangan prima
    if bilangan_40 % 2 == 0:
        return False  # bilangan genap selain 2 bukan prima
    # Langkah 2. Menghitung akar kuadrat dari bilangan_40 (dibulatkan ke bawah) dan ditambah 1
    sqrt_bilangan_40 = int(math.sqrt(bilangan_40)) + 1
    # Langkah 3. Melakukan loop dari 3 sampai sqrt_bilangan_40 untuk memeriksa bilangan ganjil agar mengurangi jumlah pembagian yang perlu diperiksa.
    for i in range(3, sqrt_bilangan_40, 2):
        # Langkah 4. Jika bilangan_40 dapat dibagi oleh i tanpa sisa, maka bilangan_40 bukan bilangan prima karena memiliki faktor lain selain 1 dan dirinya sendiri.
        if bilangan_40 % i == 0:
            return False
    # Langkah 5. Setelah melakukan pengujian, maka dapat ditunjukkan suatu bilangan prima atau bukan
    return True

# Nilai n(2 digit terakhir dari NIM)
n_40 = 40

# Menghitung batas bawah (1000n) dan batas atas (1000(n+1)) berdasarkan nilai n
batas_bawah_40 = 1000 * n_40    # Batas bawah
batas_atas_40 = 1000 * (n_40 + 1)   # Batas atas

# Mencetak bilangan prima dalam rentang tersebut
print(f"Bilangan prima antara {batas_bawah_40} dan {batas_atas_40} adalah:")
for prima_40 in range(batas_bawah_40, batas_atas_40 + 1):
    if is_prime(prima_40):
        print(prima_40)
```
