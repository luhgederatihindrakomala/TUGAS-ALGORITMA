```python
# Task 1 (Menentukan akar-akar dari persamaan ax**2+bx+c=0)
# Persamaan kuadrat dengan ax**2+bx+c=0
# Suatu persamaan dapat dicari akar-akarnya dengan cara faktorisasi, rumus abc

# Langkah 1. Akan dipanggil numpy
import numpy as np
# Langkah 2. Akan dipanggil pyplot
import matplotlib.pyplot as plt
# Langkah 3. Akan dipanggil cmath
import cmath
# Langkah 4. Masukkan a, b, c dari persamaan  ax**2+bx+c=0
a_40 = float(input("Nilai a:"))
b_40 = float(input("Nilai b: "))
c_40 = float(input("Nilai c: "))
if a_40!=0:
  print("Akan terbentuk persamaan kuadrat")
else:
  print("Tidak bisa terbentuk persamaan kuadrat")
print("Persamaan kuadrat yang terbentuk adalah ({}x**2+{}x+{})".format(a_40,b_40,c_40))
# Langkah 5. rumus diskriminan
D_40 = (b_40**2-4*a_40*c_40)
# Langkah 6. rumus X1 sebagai akar pertama,X2 sebagai akar kedua
import cmath
X1_40 = (-b_40-cmath.sqrt(D_40))/2*a_40
X2_40 = (-b_40+cmath.sqrt(D_40))/2*a_40
print("Akar-akar dari persamaan kuadrat tersebut adalah ({:.0f},{:.0f})".format(X1_40,X2_40))
```
