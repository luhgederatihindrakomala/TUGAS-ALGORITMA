```python
# Persamaan garis yang melewati 2 titik
# Garis pertama melewati (a,b) dan (c,d)
# Garis kedua melewati (e,f) dan (g,h)
# Terdapat 3 kemungkinan dari kedua persamaan garis ini:
# 1. Terdapat perpotongan di satu titik (gradien kedua garis tidak sama)
# 2. Terdapat tak berhingga banyaknya solusi (kedua garis berhimpitan) (kedua garis sejajar)
# 3. Tidak terdapat perpotongan garis (tidak terdapat solusi) (kedua garis sejajar)

import numpy as np
import matplotlib.pyplot as plt

a_40 = float(input("Nilai x dari titik pertama dari garis pertama adalah "))
b_40 = float(input("Nilai y dari titik pertama dari garis pertama adalah "))
c_40 = float(input("Nilai x dari titik kedua dari garis pertama adalah "))
d_40 = float(input("Nilai y dari titik kedua dari garis pertama adalah "))

e_40 = float(input("Nilai x dari titik pertama dari garis kedua adalah "))
f_40 = float(input("Nilai y dari titik pertama dari garis kedua adalah "))
g_40 = float(input("Nilai x dari titik kedua dari garis kedua adalah "))
h_40 = float(input("Nilai y dari titik kedua dari garis kedua adalah "))

m_1_40 = (d_40-b_40)/(c_40-a_40) # Gradien garis pertama
m_2_40 = (h_40-f_40)/(g_40-e_40) # Gradien garis kedua

# Persamaan garis pertama adalah y=m_1(x-a)+b=m_1x-m_1a+b
# Persamaan garis kedua adalah y=m_2(x-e)+f=m_2x-m_2e+f

c_1_40 = -m_1_40*a_40+b_40 # Titik potong garis pertama terhadap sumbu y
c_2_40 = -m_2_40*e_40+f_40 # titik potong garis kedua terhadap sumbu y garis kedua

print("Persamaan garis pertama adalah y={}x+{}".format(m_1_40,c_1_40))
print("Persamaan garis kedua adalah y={}x+{}".format(m_2_40,c_2_40))

if m_1_40==m_2_40:
    #print("Kedua garis sejajar sehingga kemungkinannya adalah tidak terdapat solusi atau terdapat tak berhingga banyaknya solusi")
    if c_1_40 == c_2_40:
        print("Kedua garis berhimpitan sehingga akan terdapat tak berhingga banyaknya solusi")
        x=np.linspace(-5,5,100)
        plt.plot(x,m_1_40*(x-a_40)+b_40,color='orange')
        plt.plot(x,m_2_40*(x-e_40)+f_40,color='purple')
        plt.show()
    else:
        print("Kedua garis tidak mungkin berpotongan, sehingga tidak akan terdapat solusi.")
        x=np.linspace(-5,5,100)
        plt.plot(x,m_1_40*(x-a_40)+b_40,color='orange')
        plt.plot(x,m_2_40*(x-e_40)+f_40,color='purple')
        plt.show()
else:
    print("Akan terdapat satu titik potong")
    tp_x_40 = (-m_2_40*e_40+f_40+m_1_40*a_40-b_40)/(m_1_40-m_2_40)
    tp_y_40 = m_1_40*(tp_x_40-a_40)+b_40
    print("Titik potong kedua garis adalah ({},{})".format(tp_x_40,tp_y_40))

    x=np.linspace(tp_x_40-5,tp_x_40+5,100)
    plt.plot(x,m_1_40*(x-a_40)+b_40,color='orange')
    plt.plot(x,m_2_40*(x-e_40)+f_40,color='purple')
    plt.axvline(x=0,color='grey')
    plt.axhline(y=0,color='grey')
    plt.show()
```
