```python
import numpy as np
import matplotlib.pyplot as plt

print("Kita akan membuat garis dari titik-titik yang diketahui dengan langkah akan dicari gradiennya terlebih dahulu untuk kemudian memdapatkan persamaan garisnya")
print("Sebelum itu tuliskan dulu berapa titik-titik yang nantinya akan menjadi persamaan garis")

#  Persamaan garis pertama pada titik (n_40,o_40) dan titik (p_40,q_40)
n_40 = 2 #Titik (n_40,o_40)
o_40 = -3
p_40 = 1 #Titik (p_40,q_40)
q_40 = 4

print("Akan dicari berapa m1 dari titik yang diketahui")
m1_40 = (q_40-o_40)/(p_40-n_40)
print("Gradien garis pertama adalah {}".format(m1_40))

print("Karena gradien pertama (m1) sudah didapat, maka bisa dicari persamaan garis pertama")
print("Persamaan garis pertama adalah y={}x+{}".format(m1_40,-m1_40*n_40+o_40))
#y1_40 = m1_40*(x-n_40)+o_40

# Persamaan garis kedua pada titik (r_40,s_40) dan titik (t_40,u_40)
r_40 = 6 #Titik (r_40,s_40)
s_40 = 3
t_40 = 2 #Titik (t_40,u_40)
u_40 = 7

print("Akan dicari berapa m2 dari titik yang diketahui")
m2_40 = (u_40-s_40)/(t_40-r_40)
print("Gradien garis kedua adalah {}".format(m2_40))

print("Karena gradien kedua (m2) sudah didapat, maka bisa dicari persamaan garis kedua")
print("Persamaan garis kedua adalah y={}x+{}".format(m2_40,-m2_40*r_40+s_40))
#y2_40 = m2_40*(x-r_40)+s_40

print("Perhatikan m1 dan m2 apakah sama atau tidak")
print("Jika m1=m2, maka persamaan garisnya tidak memiliki solusi. Jika lainnya, maka akan dicari titik potongnya.")
if m1_40==m2_40:
  print("Ternyata gradiennya sama. Oleh karena itu, persamaan garis ini tidak memiliki solusi.")
else:
  # Titik potong kedua garis tersebut terletak pada (tp_x_40,tp_y_40)
  tp_x_40 = (m1_40*n_40-o_40-m2_40*r_40+s_40)/(m1_40-m2_40)
  tp_y_40 = m2_40*(tp_x_40-r_40)+s_40
  print("Titik potong kedua garis tersebut terletak pada ({},{})".format(tp_x_40,tp_y_40))

  x_40 = np.arange(tp_x_40-10,tp_y_40+10,0.01)
  plt.figure(figsize=(4,4),dpi=500)

  print("Perhatikan gambar, persamaan garis pertama berwarna hijau dan persamaan garis kedua berwarna kuning")
  plt.plot(x_40,m1_40*(x_40-n_40)+o_40,color='green',label='y={}x+{}'.format(m1_40,-m1_40*n_40+o_40))
  plt.plot(x_40,m2_40*(x_40-r_40)+s_40,color='yellow',label='y={}x+{}'.format(m2_40,-m2_40*r_40+s_40))
  plt.plot(tp_x_40,tp_y_40,color='red',marker='o',markersize=5)
  plt.legend(loc='best')
  plt.show()

print("Maka, gambarnya tampak pada layar")
```
